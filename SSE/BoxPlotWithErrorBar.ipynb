```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# f1
labels = ['Random Forest', 'Extra Trees','MLP', 'KNN', 'HMM']
values = [0.936949165,0.942849269,0.949150972,0.923292112,0.888214409]
stddev = [0.07932096,0.070128975,0.047493873,0.064280534,0.085330157]
x_pos = np.arange(len(labels))
fig, ax = plt.subplots()
frontsize = 20
#fig = matplotlib.pyplot.gcf()
fig.set_size_inches(18.5, 10.5)
#fig.savefig('test2png.png', dpi=100)
ax.bar(x_pos, values, yerr=stddev, align='center', alpha=0.5, ecolor='black', capsize=10)
ax.set_ylabel('F1', fontsize=frontsize)
ax.set_xticks(x_pos)
ax.set_xticklabels(labels, fontsize=frontsize)
#ax.set_title('', fontsize=frontsize)
ax.yaxis.grid(True)
plt.ylim(0, 1)

# Save the figure and show
plt.tight_layout()
plt.savefig('F1.png', dpi=100)
plt.show()
```

```python
#  acc
labels = ['Random Forest', 'Extra Trees','MLP', 'KNN', 'HMM']
values = [0.934931733,0.942431003,0.951772102,0.920818626,0.89359004]
stddev = [0.086196673,0.074788064,0.04011034,0.068820619,0.076447543]

x_pos = np.arange(len(labels))
fig, ax = plt.subplots()
frontsize = 20
#fig = matplotlib.pyplot.gcf()
fig.set_size_inches(18.5, 10.5)
#fig.savefig('test2png.png', dpi=100)
ax.bar(x_pos, values, yerr=stddev, align='center', alpha=0.5, ecolor='black', capsize=10)
ax.set_ylabel('Accuracy', fontsize=frontsize)
ax.set_xticks(x_pos)
ax.set_xticklabels(labels, fontsize=frontsize)
#ax.set_title('Accuracy', fontsize=frontsize)
ax.yaxis.grid(True)
plt.ylim(0, 1)

# Save the figure and show
plt.tight_layout()
plt.savefig('Accuracy.png', dpi=100)
plt.show()
```
